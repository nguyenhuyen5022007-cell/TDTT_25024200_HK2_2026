### Bài 4: 
Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím

```python
h = int(input())
for i in range(1,h+1):
    spaces = " " * (h - i)
    stars = "*" * (2 * i - 1)
    print(spaces + stars)
```

```python
h = int(input("Nhập độ cao h: "))
for i in range(1, h + 1):
    spaces = " " * (h - i)
    if i == 1:
        print(spaces + "*")
    elif i == h:
        print("*" * (2 * h - 1))
    else:
        inner_spaces = " " * (2 * i - 3)
        print(spaces + "*" + inner_spaces + "*")
```

### Bài 7: 
Nhập vào số nguyên dương a, đếm số ước dương của a

```python
a = int(input())
dem = 0
for i in range (1,a+1):
    if a%i==0:
        dem += 1
        if i*i != a:
            dem += 1
print(dem)
```

```python
import math

# Nhập số nguyên dương a
a = int(input())
dem = 0
# Duyệt từ 1 đến căn bậc hai của a
for i in range(1, int(math.sqrt(a)) + 1):
    if a % i == 0:
        dem += 1  # i là ước thứ nhất
        # Tìm ước còn lại là a // i
        if i * i != a:
            dem += 1 # Cộng thêm ước thứ hai nếu nó khác i

print(dem)
```

### Bài 9: 
Nhập vào số nguyên dương a, kiểm tra xem a có phải là số chính phương hay không? Số chính phương là số có căn bậc hai là một số nguyên VD: 1, 4, 9, 16, 25….

```python
import math

a = int(input("Nhập vào số nguyên dương a: "))

if a < 0:
    print(f"{a} không phải là số chính phương (số âm).")
else:
    can_bac_hai = math.isqrt(a)  

    if can_bac_hai * can_bac_hai == a:
        print(f"{a} là số chính phương.")
    else:
        print(f"{a} không phải là số chính phương.")
```

### Bài 10: 
Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không.
Số hoàn hảo là số có tổng các ước (ước nhỏ hơn nó) bằng chính nó, VD: số 6, 28…

```python
a = int(input())

if a <= 0:
    print("Vui lòng nhập một số nguyên dương!")
else:
    tong_uoc = 0
    for i in range(1, a):
        if a % i == 0:
            tong_uoc += i
        
    if tong_uoc == a:
        print(f"{a} là số hoàn hảo.")
    else:
        print(f"{a} không phải là số hoàn hảo.")
```

### Bài 12: 
Nhập số nguyên dương n, Cho S(k) = 1 + 2 + 3 + … + k. Tìm k sao cho S(k)
lớn nhất nhưng nhỏ hơn n 


```python
n = int(input())

s = 0  
k = 0  

while s + (k + 1) < n:
    k += 1    
    s += k  

print(f"Số k lớn nhất để S(k) < {n} là: k = {k}")
print(f"Khi đó, S({k}) = {s}")
```

### Bài 13: 
Nhập vào số nguyên dương A, tìm n nhỏ nhất sao cho
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A

```python
A = int(input())

tong_S = 0.0
n = 0

while tong_S <= A:
    n += 1
    tong_S += 1 / n

print(f"Số n nhỏ nhất tìm được là: {n}")
print(f"Tổng S({n}) xấp xỉ: {tong_S:.5f}")
```

### Bài 14: 
Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập
xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng

```python
n = int(input())

if n == -1:
    print("Bạn đã dừng nhập ngay từ đầu.")
else:
    max_val = n
    min_val = n
    
    while True:
        n = int(input())

        if n == -1:
            break
        if n > max_val:
            max_val = n
        if n < min_val:
            min_val = n

    print("-" * 20)
    print(f"Số lớn nhất là: {max_val}")
    print(f"Số nhỏ nhất là: {min_val}")
```

### Bài 15: 
Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số. Làm bằng nhiều
cách nhất có thể, ko dùng mảng

```python
n = int(input())
count = 0

if n == 0:
    count = 1
else:
    temp = abs(n) 
    while temp > 0:
        temp //= 10 
        count += 1

print(f"Số chữ số là: {count}")
```

### Bài 16: 
Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu
chữ số lẻ 


```python
n = int(input())

dem_chan = 0
dem_le = 0

if n == 0:
    dem_chan = 1
else:
    temp = abs(n) 
    while temp > 0:
        chu_so = temp % 10 
        
        if chu_so % 2 == 0:
            dem_chan += 1
        else:
            dem_le += 1
            
        temp //= 10 

print(f"Số chữ số chẵn: {dem_chan}")
print(f"Số chữ số lẻ: {dem_le}")
```

### Bài 17: 
Nhập vào số nguyên dương n, tính tổng các chữ số của n 


```python
n = int(input("Nhập vào số nguyên dương n: "))

tong_chu_so = 0
temp = abs(n)

while temp > 0:
    chu_so_cuoi = temp % 10    
    tong_chu_so += chu_so_cuoi 
    temp //= 10                

print(f"Tổng các chữ số của {n} là: {tong_chu_so}")
```

### Bài 18: 
Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 3^k hay
không 


```python
n = int(input())

if n <= 0:
    print(f"{n} không phải là số dạng 3^k")
else:
    temp = n
    while temp % 3 == 0:
        temp //= 3
    if temp == 1:
        print(f"{n} là số dạng 3^k")
    else:
        print(f"{n} không phải là số dạng 3^k")
```

### Bài 19: 
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong
dãy số fibonacci lớn nhất nhưng không vượt quá A 

```python
A = int(input())

if A < 1:
    print("Không có số Fibonacci nào thỏa mãn.")
elif A == 1:
    print("Số Fibonacci lớn nhất không vượt quá A là: 1")
else:
    f1 = 1
    f2 = 1
    f3 = f1 + f2 
    
    while f3 <= A:
        f1 = f2
        f2 = f3
        f3 = f1 + f2
    
    print(f"Số Fibonacci lớn nhất không vượt quá {A} là: {f2}")
```

### Bài 20: 
Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố. 

```python
n = int(input())

dem = 0
temp = abs(n)

if n == 0:
    dem = 0 
else:
    while temp > 0:
        chu_so = temp % 10  

        if chu_so == 2 or chu_so == 3 or chu_so == 5 or chu_so == 7:
            dem += 1
            
        temp //= 10 

print(f"Số lượng chữ số là số nguyên tố trong {n} là: {dem}")
```
