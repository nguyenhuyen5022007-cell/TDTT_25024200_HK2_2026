### W4A2 [PrimeNumber]

Kiểm tra số này có phải là số nguyên tố hay không?

```python
n = int(input())
if n<=1:
    print(f"{n} khong la so nguyen to")
else:
    c = 0
    for i in range (1, n+1):
        if n%i==0:
            c += 1
    if c == 2:
        print(f"{n} la so nguyen to")
    else:
        print(f"{n} khong la so nguyen to")
```

### W4A3 [Factorial]
Viết chương trình tính giai thừa của một số nguyên n cho trước (n!=1×2×…×n)
với 0 < n < 100. 


```python
def tinh_giai_thua(n):
    giai_thua = 1
    # Vòng lặp từ 1 đến n
    for i in range(1, n + 1):
        giai_thua *= i
    return giai_thua

# Nhập dữ liệu
n = int(input("Nhập số nguyên n (0 < n < 100): "))

if 0 < n < 100:
    print(f"{n}! = {tinh_giai_thua(n)}")
else:
    print("Vui lòng nhập n trong khoảng (0, 100)")
```

### W4A6
"[SumPrimesInRange]
Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn
[a, b]. Mỗi dòng là một bộ test (a b)."


```python
import sys

def is_prime(n):
    """Hàm kiểm tra số nguyên tố"""
    if n < 2:
        return False
    if n == 2 or n == 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False

    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

def solve():
    for line in sys.stdin:
        if not line.strip():
            continue
        try:
            a, b = map(int, line.split())
            
            total_sum = 0
            for num in range(a, b + 1):
                if is_prime(num):
                    total_sum += num
            
            print(total_sum)
        except ValueError:
            break

if __name__ == "__main__":
    solve()
```

### W4A8
ReverseAddPalindrome
Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số  đối xứng (palindrome). In ra số bước và giá trị palindrome thu được

```python
def is_palindrome(n):
    """Kiểm tra xem số n có phải số đối xứng không"""
    s = str(n)
    return s == s[::-1]

def reverse_num(n):
    """Đảo ngược các chữ số của n"""
    return int(str(n)[::-1])

n = int(input("Nhập n: "))

steps = 0

while not is_palindrome(n):
    n = n + reverse_num(n)
    steps += 1

print(f"{steps} {n}")
```

### W4A10
"[LongestCollatzUpToN]
Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất
và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số
nhỏ nhất." 


```python
def get_collatz_length(num, memo):
    """Tính độ dài dãy Collatz với kỹ thuật ghi nhớ (memoization)"""
    original_num = num
    length = 1
    temp_path = []
    
    while num > 1:
        if num in memo:
            length += memo[num] - 1
            break
        
        temp_path.append(num)
        if num % 2 == 0:
            num //= 2
        else:
            num = 3 * num + 1
        length += 1

    curr_len = length
    for p in temp_path:
        if p not in memo:
            memo[p] = curr_len
        curr_len -= 1
        
    return length

def solve():
    try:
        n = int(input())
        
        max_length = 0
        best_x = 1
        memo = {1: 1} 

        for i in range(1, n + 1):
            curr_length = get_collatz_length(i, memo)

            if curr_length > max_length:
                max_length = curr_length
                best_x = i

        print(f"{best_x} {max_length}")
        
    except ValueError:
        pass

if __name__ == "__main__":
    solve()
```

### W4A11
[EvenDivisors]
Viết chương trình đếm số lượng ước số chẵn của số nguyên dương n với n < 10^6

```python
import math

def count_even_divisors(n):
    if n % 2 != 0:
        return 0
    
    count = 0
    square_root = int(math.sqrt(n))
    
    for i in range(1, square_root + 1):
        if n % i == 0:
            if i % 2 == 0:
                count += 1
            if (n // i) % 2 == 0 and (n // i) != i:
                count += 1
                
    return count

try:
    n = int(input("Nhập số nguyên dương n: "))
    if n <= 0:
        print("Vui lòng nhập số nguyên dương!")
    else:
        result = count_even_divisors(n)
        print(f"Số lượng ước số chẵn của {n} là: {result}")
except ValueError:
    print("Vui lòng nhập một số nguyên hợp lệ.")
```

Ví dụ 220 và 284 là một cặp số thân thiết

110. Tổng của 11 ước này cho chúng ta 284
284 ngoài bản thân nó, còn có 5 ước số khác là: 1, 2, 4, 71, 142. Tổng của chúng
vừa đúng bằng 220

Nếu có, in ra màn hình "true", ngược lại, in ra "false" 


```python
m,n = map(int, input().split())
tong_uoc_m = 0
for i in range (1,m):
    if m%i==0:
        tong_uoc_m += i
tong_uoc_n = 0
for i in range (1,n):
    if n%i==0:
        tong_uoc_n += i
if tong_uoc_m == n and tong_uoc_n == m:
    print("true")
else:
    print("false")
```
