```python
import math # Thư viện này dùng để tính căn bậc hai (sqrt)

print("--- CHƯƠNG TRÌNH GIẢI PHƯƠNG TRÌNH BẬC 2: ax^2 + bx + c = 0 ---")

# Bước 1: Nhập các hệ số từ bàn phím
# float giúp máy tính hiểu đây là số thập phân
a = float(input("Nhập hệ số a (a khác 0): "))
b = float(input("Nhập hệ số b: "))
c = float(input("Nhập hệ số c: "))

# Kiểm tra điều kiện a phải khác 0
if a == 0:
    print("Đây không phải là phương trình bậc 2!")
else:
    # Bước 2: Tính Delta (b bình phương trừ 4ac)
    delta = b**2 - 4*a*c
    print(f"-> Delta tính được là: {delta}")

    # Bước 3: Xét các trường hợp của Delta
    if delta < 0:
        print("Kết luận: Phương trình vô nghiệm.")
        
    elif delta == 0:
        x = -b / (2 * a)
        print(f"Kết luận: Phương trình có nghiệm kép x1 = x2 = {x}")
        
    else:
        # Tính căn bậc hai của Delta
        can_delta = math.sqrt(delta)
        
        # Áp dụng công thức nghiệm
        x1 = (-b + can_delta) / (2 * a)
        x2 = (-b - can_delta) / (2 * a)
        
        print(f"Kết luận: Phương trình có 2 nghiệm phân biệt:")
        print(f"x1 = {x1}")
        print(f"x2 = {x2}")
```

```python
# Bước 1: Nhập hai số nguyên a và b trên cùng một dòng
data = input().split()
a = int(data[0])
b = int(data[1])

# Bước 2: Tính toán các giá trị
tong = a + b
hieu = a - b
tich = a * b
phan_nguyen = a // b  # Phép chia lấy phần nguyên
phan_du = a % b       # Phép chia lấy phần dư
chia_thuc = a / b     # Phép chia số thực

# Bước 3: In kết quả theo đúng định dạng
print(f"Tổng: {tong}")
print(f"Hiệu: {hieu}")
print(f"Tích: {tich}")
print(f"Phần nguyên: {phan_nguyen}")
print(f"Phần dư: {phan_du}")
print(f"Chia thực: {chia_thuc:.2f}")
```
