### Bài 1: 
Nhập vào 2 cạnh của một hình chữ nhật. In ra màn hình diện tích và chu vi của nó.

```python
# Nhập chiều dài và chiều rộng
a = float(input("Nhập chiều dài: "))
b = float(input("Nhập chiều rộng: "))

# Tính toán
dien_tich = a * b
chu_vi = (a + b) * 2

# In kết quả
print(f"Diện tích hình chữ nhật là: {dien_tich}")
print(f"Chu vi hình chữ nhật là: {chu_vi}")
```

### Bài 2: 
Nhập vào bán kính của hình tròn. In ra màn hình diện tích và chu vi của
nó.

```python
import math

# Nhập bán kính
r = float(input("Nhập bán kính hình tròn: "))

# Tính toán (dùng math.pi để có độ chính xác cao hơn)
dien_tich = math.pi * r**2
chu_vi = 2 * math.pi * r

# In kết quả (làm tròn đến 2 chữ số thập phân)
print(f"Diện tích hình tròn là: {round(dien_tich, 2)}")
print(f"Chu vi hình tròn là: {round(chu_vi, 2)}")
```

### Bài 3 : 
Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là
độ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh
cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân,
Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn
hình.

```python
import math

# Nhập 3 cạnh
a = float(input("Nhập cạnh a: "))
b = float(input("Nhập cạnh b: "))
c = float(input("Nhập cạnh c: "))

# 1. Kiểm tra điều kiện tam giác
if a + b > c and a + c > b and b + c > a:
    print("Đây là ba cạnh của một tam giác.")
    
    # 2. Phân loại tam giác
    loai_tg = ""
    if a == b == c:
        loai_tg = "Đều"
    elif a == b or b == c or a == c:
        if (a**2 + b**2 == c**2) or (a**2 + c**2 == b**2) or (b**2 + c**2 == a**2):
            loai_tg = "Vuông Cân"
        else:
            loai_tg = "Cân"
    elif (a**2 + b**2 == round(c**2, 2)) or (a**2 + c**2 == round(b**2, 2)) or (b**2 + c**2 == round(a**2, 2)):
        loai_tg = "Vuông"
    else:
        loai_tg = "Thường"
    
    print(f"Kiểu tam giác: Tam giác {loai_tg}")

    # 3. Tính chu vi và diện tích
    chu_vi = a + b + c
    p = chu_vi / 2 # Nửa chu vi
    dien_tich = math.sqrt(p * (p - a) * (p - b) * (p - c))
    
    print(f"Chu vi: {chu_vi}")
    print(f"Diện tích: {round(dien_tich, 2)}")

else:
    print("Khong la ba canh cua mot tam giac")
```

### Bài 5: 
Viết chương trình giải phương trình bậc 2 (ax^2 + bx + c= 0)

```python
import math

print("Giải phương trình bậc 2: ax^2 + bx + c = 0")
a = float(input("Nhập a: "))
b = float(input("Nhập b: "))
c = float(input("Nhập c: "))

if a == 0:
    # Trở về bài toán phương trình bậc 1
    if b == 0:
        print("Vô số nghiệm" if c == 0 else "Vô nghiệm")
    else:
        print(f"Phương trình có 1 nghiệm: x = {-c/b}")
else:
    delta = b**2 - 4*a*c
    print(f"Delta = {delta}")
    
    if delta < 0:
        print("Phương trình vô nghiệm thực.")
    elif delta == 0:
        x = -b / (2 * a)
        print(f"Phương trình có nghiệm kép: x1 = x2 = {x}")
    else:
        x1 = (-b + math.sqrt(delta)) / (2 * a)
        x2 = (-b - math.sqrt(delta)) / (2 * a)
        print(f"Phương trình có 2 nghiệm phân biệt:")
        print(f"x1 = {round(x1, 2)}")
        print(f"x2 = {round(x2, 2)}")
```

### Bài 6: 
Tìm giá trị lớn nhất trong 4 số a, b, c, d (a, b, c, được nhập từ bàn phím).

```python
# Nhập dữ liệu
a = float(input("Nhập số thứ nhất (a): "))
b = float(input("Nhập số thứ hai (b): "))
c = float(input("Nhập số thứ ba (c): "))
d = float(input("Nhập số thứ tư (d): "))

# --- Bài 6: Tìm Max ---
max_val = a  # Giả sử a là lớn nhất
if b > max_val: max_val = b
if c > max_val: max_val = c
if d > max_val: max_val = d

# In kết quả
print("-" * 20)
print(f"Giá trị lớn nhất là: {max_val}")
```

### Bài 8: 
Giải hệ phương trình tuyến tính:

ax+by = m

cx+dy= n

```python
print("Giải hệ phương trình: ax + by = m và cx + dy = n")
a = float(input("Nhập a: "))
b = float(input("Nhập b: "))
m = float(input("Nhập m: "))
c = float(input("Nhập c: "))
d = float(input("Nhập d: "))
n = float(input("Nhập n: "))

D = a * d - b * c
Dx = m * d - b * n
Dy = a * n - m * c

if D != 0:
    x = Dx / D
    y = Dy / D
    print(f"Hệ có nghiệm duy nhất: x = {x}, y = {y}")
else:
    if Dx == 0 and Dy == 0:
        print("Hệ phương trình có vô số nghiệm.")
    else:
        print("Hệ phương trình vô nghiệm.")
```

### Bài 9: 
Nhập vào thời gian 1 công việc nào đó là x giây. Hãy chuyển đổi và viết
ra màn hình số thời gian trên dưới dạng bằng bao nhiêu giờ, bao nhiêu phút, bao

### Bài 10: 
Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào một
điểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
import math

x0 = float(input("Nhập tọa độ x tâm I: "))
y0 = float(input("Nhập tọa độ y tâm I: "))
r = float(input("Nhập bán kính R: "))

xa = float(input("Nhập tọa độ x điểm A: "))
ya = float(input("Nhập tọa độ y điểm A: "))

# Tính khoảng cách từ tâm đến điểm A
khoang_cach = math.sqrt((xa - x0)**2 + (ya - y0)**2)

# Kiểm tra (Dùng round để tránh sai số số thực)
if round(khoang_cach, 6) == round(r, 6):
    print("Điểm A thuộc đường tròn.")
elif khoang_cach < r:
    print("Điểm A nằm trong đường tròn.")
else:
    print("Điểm A nằm ngoài đường tròn.")
```
