### Bài 2: 
Một khu đất hình chữ nhật có chiều rộng a, chiều dài b. Người ra xây một khu
vui chơi hình tròn tại vị trí trung tâm khu đất như hình vẽ. Nhập vào a, b. Hãy tính phần
diện tích còn để trồng cây xung quanh là bao nhiêu. Biết pi = 3.14, lấy kết quả đến 2
chữ số phần thập phân. 


```python
# Nhập chiều rộng a và chiều dài b
a = float(input("Nhập chiều rộng a: "))
b = float(input("Nhập chiều dài b: "))

# 1. Tính diện tích hình chữ nhật
s_hcn = a * b

# 2. Xác định bán kính hình tròn (r)
# Giả sử hình tròn tiếp xúc với cạnh ngắn hơn, ở đây là a
r = a / 2

# 3. Tính diện tích hình tròn (pi = 3.14)
s_tron = 3.14 * (r ** 2)

# 4. Tính diện tích còn lại để trồng cây
s_con_lai = s_hcn - s_tron

# In kết quả lấy 2 chữ số thập phân
print(f"Diện tích phần trồng cây là: {s_con_lai:.2f}")
```

### Bài 3:
Viết chương trình nhập vào một “kí tự” c, nếu nó là chữ hoa thì in ra chữ thường
tương ứng và ngược lại nếu nó là chữ thường thì in ra chữ hoa tương ứng của nó. (xem
trước cấu trúc lựa chọn: if….else, ma) 

```python
# Nhập ký tự từ bàn phím
c = input("Nhập vào một kí tự: ")

# Sử dụng cấu trúc if...else để lựa chọn
if c.isupper():
    # Nếu là chữ hoa, chuyển thành chữ thường
    ket_qua = c.lower()
    print(f"Ký tự thường tương ứng là: {ket_qua}")
elif c.islower():
    # Nếu là chữ thường, chuyển thành chữ hoa
    ket_qua = c.upper()
    print(f"Ký tự hoa tương ứng là: {ket_qua}")
else:
    # Trường hợp người dùng nhập số hoặc ký hiệu đặc biệt
    print("Đây không phải là chữ cái!")
```

### Bài 4: 
Viết chương trình nhập vào một “kí tự”, Kiểm tra xem c có phải là kí tự chữ cái
không (hoa hoặc thường). Nếu đúng thì in ra “{c} là kí tu alphabet”, nếu sai in ra “{c}
không phải là kí tự alphabet. 


```python
c = input("Nhập vào một kí tự: ")

if c.isalpha():
    print(f"{c} là kí tự alphabet")
else:
    print(f"{c} không phải là kí tự alphabet")
```

### Bài 5: 
Nhập vào chữ cái hoa A, in ra chữ cái thường liền trước chữ cái a thường tương
ứng của A. Xét TH khi c = ‘A’ là đặc biệt. 


```python
A_hoa = input("Nhập vào chữ cái hoa: ")

# Chuyển về chữ thường tương ứng
a_thuong = A_hoa.lower()

if A_hoa == 'A':
    print("Trường hợp đặc biệt: Không có chữ cái thường nào đứng trước 'a'")
else:
    # Lấy mã ASCII của a_thuong, trừ đi 1 để lấy mã của chữ đứng trước
    ma_truoc = ord(a_thuong) - 1
    # Chuyển mã đó về lại dạng ký tự
    chu_truoc = chr(ma_truoc)
    print(f"Chữ cái thường liền trước là: {chu_truoc}")
```

### Bài 6. 
Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không,
nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó
không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”. 


```python
import math # Thư viện để dùng hàm căn bậc hai sqrt()

# Nhập 3 cạnh
a = float(input("Nhập cạnh a: "))
b = float(input("Nhập cạnh b: "))
c = float(input("Nhập cạnh c: "))

# 1. Kiểm tra điều kiện tồn tại tam giác
if a + b > c and a + c > b + b + c > a:
    # 2. Nếu đúng, tính diện tích bằng công thức Heron
    p = (a + b + c) / 2
    s = math.sqrt(p * (p - a) * (p - b) * (p - c))
    
    print(f"Diện tích tam giác là: {s:.1f}")
else:
    # 3. Nếu sai
    print("Khong phai 3 canh cua tam giac")
```

### Bài 7: 
Nhập vào một chuỗi gồm các kí tự thường độ dài >=20, in ra chữ cái thứ năm và
thứ chín của chuỗi này. 


```python
# Nhập chuỗi
chuoi = input("Nhập vào một chuỗi (độ dài >= 20): ")

# Kiểm tra độ dài trước khi in để đảm bảo không bị lỗi
if len(chuoi) >= 20:
    chu_thu_5 = chuoi[4] # Vị trí thứ 5 là index 4
    chu_thu_9 = chuoi[8] # Vị trí thứ 9 là index 8
    
    print(f"Chữ cái thứ năm là: {chu_thu_5}")
    print(f"Chữ cái thứ chín là: {chu_thu_9}")
else:
    print("Chuỗi bạn nhập quá ngắn, vui lòng nhập ít nhất 20 ký tự!")
```
