### W5A6 
Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!) 


```python
def factorial(n):
    gt = 1
    if n >= 0:
        for i in range(1, n+1):
            gt *= i
        return gt
    return "None"
    
assert factorial(0) == 1
assert factorial(3) == 6
assert factorial(5) == 120
assert factorial(1) == 1
assert factorial(-10) == "None"

print("pass all test")      
```

### W5A10
Hai từ được gọi là đẳng cấu nếu các kí tự trong một từ có thể được ánh xạ để được từ còn lại. Ánh
xạ một kí tự nghĩa là thay đổi toàn bộ các kí tự đó trong từ bởi một kí tự khác. Thứ tự của các kí tự
được giữ không đổi. Hai kí tự khác nhau không thể ánh xạ đến cùng một kí tự, nhưng một kí tự có
thể ánh xạ thành chính nó. Ví dụ, từ "abca" và "zbxz" là hai từ đẳng cấu vì ta có thể ánh xạ 'a' thành
'z', 'b' thành 'b' và 'c' thành 'x'.
Viết hàm trả về \(true\) nếu hai từ \(a\) và \(b\) là đẳng cấu, ngược lại \(false\). Bạn có thể giả sử hai
từ đầu vào có cùng độ dài. 


```python
def is_isomorphic(s: str, t: str) -> bool:
    map_s_to_t = {}
    map_t_to_s = {}

    for char_s, char_t in zip(s, t):
        if char_s in map_s_to_t:
            if map_s_to_t[char_s] != char_t:
                return False
        else:
            map_s_to_t[char_s] = char_t

        if char_t in map_t_to_s:
            if map_t_to_s[char_t] != char_s:
                return False
        else:
            map_t_to_s[char_t] = char_s

    return True

s,t = map(str, input().split())
print(is_isomorphic(s,t))
```

### W5A12
Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) sao cho dividend = divisor*q + r và 0 <= |r| < |divisor|.

```python
def divide_with_remainder(dividend, divisor):
    if divisor == 0:
        return False
    q = dividend//divisor
    r = dividend%divisor
    return q,r
dividend, divisor = map(int, input().split())
print(divide_with_remainder(dividend, divisor))
        
```

### W5A14 
Viết hàm compute(a, b, op) thực hiện phép tính a (+|-|*|/) b tùy theo ký tự op. 

```python
def compute(a, b, op):
    if op == '+':
        return a+b
    elif op == '-':
        return a-b
    elif op == '*':
        return a*b
    elif op == '/':
        if b==0:
            return False
        else:
            return a/b
    else:
        return False
a,b = map(int, input().split())
op = str(input())
print(compute(a, b, op))
```

### W5A15 
Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông.

```python
def is_right_triangle(a, b, c):
    if a+b>c and b+c>a and a+c>b:
        if a**2+b**2==c**2 or a**2+c**2==b**2 or c**2+b**2==a**2:
            return True
    return False
a,b,c = map(int, input().split())
print(is_right_triangle(a, b, c))
    
```
