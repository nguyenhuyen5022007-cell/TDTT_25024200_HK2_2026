### W4A18 
Nhập vào số nguyên dương a và b, in toàn bộ ước chung của a và b 


```python
import math

a = int(input("Nhập số nguyên dương a: "))
b = int(input("Nhập số nguyên dương b: "))

ucln = math.gcd(a, b)

print(f"Các ước chung của {a} và {b} là:")

for i in range(1, ucln + 1):
    if ucln % i == 0:
        print(i, end=" ")
```

### W4A20 
Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không

```python
n = int(input())
temp = n

if n <= 0:
    print(f"{n} không phải là số dạng 2^k")
else:
    while n % 2 == 0:
        n //= 2
    
    if n == 1:
        print(f"{temp} là số dạng 2^k")
    else:
        print(f"{temp} không phải là số dạng 2^k")
```

### W4A21 
Nhập vào số nguyên dương n, tính tổng các chữ số của n


```python
n = int(input("Nhập số nguyên dương n: "))
tong = 0
temp = n 

while n > 0:
    chu_so = n % 10
    tong += chu_so
    n //= 10

print(f"Tổng các chữ số của {temp} là: {tong}")
```

### W4A22 
Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ 


```python
n = int(input())
dem = 0
temp = n

if n == 0: 
    dem = 1
else:
    while n > 0:
        dem += 1
        n //= 10

print(f"Số {temp} có {dem} chữ số.")
```

### W4A23
Nhập n
Cho S(k) = 1 + 2 + 3 + … + k
Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n

```python
n = int(input())

s = 0
k = 0

while s + (k + 1) < n:
    k += 1
    s += k

print(f"Giá trị k lớn nhất là: {k}")
print(f"Tổng S({k}) = {s} (nhỏ hơn {n})")
```

### W4A25

Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.

Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python
so_nhap = int(input())

if so_nhap == -1:
    print("Bạn chưa nhập số nào hợp lệ.")
else:
    max_val = so_nhap
    min_val = so_nhap

    while True:
        n = int(input())

        if n == -1:
            break
 
        if n > max_val:
            max_val = n

        if n < min_val:
            min_val = n

    print("--- Kết quả ---")
    print(f"Số lớn nhất là: {max_val}")
    print(f"Số nhỏ nhất là: {min_val}")
```

### W4A26
[Fibonaci1]
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ
bằng tổng hai số trước đó
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt
quá A 


```python
A = int(input())

# Khởi tạo hai số đầu tiên của dãy
f1, f2 = 1, 1

if A < 1:
    print("Không có số Fibonacci nào thỏa mãn.")
elif A == 1:
    print("Số Fibonacci lớn nhất không vượt quá 1 là: 1")
else:
    f_next = f1 + f2
 
    while f_next <= A:
        f1 = f2
        f2 = f_next
        f_next = f1 + f2
  
    print(f"Số Fibonacci lớn nhất không vượt quá {A} là: {f2}")
```

### W4A27

Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu từ (quy định là chuỗi không có ký tự đặc biệt, không số, không có dấu câu, chỉ có ký tự chữ và khoảng trắng)

Dùng cách: for trên index; KHÔNG dùng hàm split(), len() 

```python
chuoi = input()

do_dai = 0
for ky_tu in chuoi:
    do_dai += 1

so_tu = 0
dang_trong_tu = False

for i in range(do_dai):
    if chuoi[i] != ' ':
        if not dang_trong_tu:
            so_tu += 1
            dang_trong_tu = True
    else:
        dang_trong_tu = False

print("Số lượng từ trong chuỗi là:", so_tu)
```

### W4A28

Nhập vào một chuỗi, hãy in từ đầu tiên trong chuỗi  Dùng cách: for trên index; KHÔNG dùng hàm split()

```python
chuoi = input()

do_dai = 0
for ky_tu in chuoi:
    do_dai += 1

index_bat_dau = -1
index_ket_thuc = do_dai 

for i in range(do_dai):
    if chuoi[i] != ' ':
        index_bat_dau = i
        break

if index_bat_dau == -1:
    print("Chuỗi chỉ toàn khoảng trắng hoặc rỗng.")
else:
    for i in range(index_bat_dau, do_dai):
        if chuoi[i] == ' ':
            index_ket_thuc = i
            break
   
    tu_dau_tien = ""
    for i in range(index_bat_dau, index_ket_thuc):
        tu_dau_tien += chuoi[i]
        
    print("Từ đầu tiên trong chuỗi là:", tu_dau_tien)
```

### W4A29
Cho một chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa, hãy đổi ký tự hoa thành thường và thường thành hoa. 

```python
chuoi = input()
ket_qua = ""

for ky_tu in chuoi:
    if ky_tu.isupper():
        ket_qua += ky_tu.lower()
    elif ky_tu.islower():
        ket_qua += ky_tu.upper()
    else:
        ket_qua += ky_tu

print("Chuỗi sau khi chuyển đổi:", ket_qua)
```

### W4A30
Nhập vào một chuỗi có dạng 3 số nguyên, mỗi số nguyên cách nhau một dấu phẩy, hãy tính
tổng 3 số nguyên đó
VD:
Nhập: 3, 12, 15
Tổng: 30 


```python
chuoi = input()
tong = 0
so_hien_tai = ""

for i in range(len(chuoi)):
    char = chuoi[i]
    if char.isdigit(): 
        so_hien_tai += char
    
    if char == "," or i == len(chuoi) - 1:
        if so_hien_tai != "":
            tong += int(so_hien_tai)
            so_hien_tai = ""

print("Tổng:", tong)
```

### W4A31
Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in
thường, bao nhiêu ký tự số

```python
chuoi = input()

dem_hoa = 0
dem_thuong = 0
dem_so = 0

for ky_tu in chuoi:
    if ky_tu.isupper():
        dem_hoa += 1
    elif ky_tu.islower():
        dem_thuong += 1
    elif ky_tu.isdigit():
        dem_so += 1

print(f"Số ký tự in hoa: {dem_hoa}")
print(f"Số ký tự in thường: {dem_thuong}")
print(f"Số ký tự số: {dem_so}")
```
